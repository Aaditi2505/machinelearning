```python
from IPython import get_ipython
from IPython.display import display
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.semi_supervised import SelfTrainingClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np

df = pd.read_csv("/content/house_prices.csv")

df = df[["area", "rooms", "location", "age", "price"]].dropna()

le = LabelEncoder()
# It seems 'rooms' is treated as a categorical feature and then modified to be a numerical target.
# If 'rooms' is indeed categorical, LabelEncoding is appropriate.
# However, the SelfTrainingClassifier is typically used with a numerical or integer target.
# The LabelEncoder will produce integers. The -1 is then introduced for unlabeled data.
df["rooms"] = le.fit_transform(df["rooms"])

np.random.seed(42)
mask = np.random.rand(len(df)) < 0.5
df.loc[mask, "rooms"] = -1

X = df.drop(["rooms", "location"], axis=1)
y = df["rooms"]

# Explicitly drop rows with NaN values from X before fitting
X = X.dropna()
# Also drop corresponding rows from y to maintain consistency
y = y[X.index]

base_model = LogisticRegression()

self_training_model = SelfTrainingClassifier(base_model, criterion='k_best')

# Check if X and y have the same number of samples after dropping NaNs
if len(X) == len(y):
    self_training_model.fit(X, y)
else:
    print("Mismatch between X and y lengths after dropping NaNs. Cannot fit the model.")


df["Predicted_Label"] = self_training_model.predict(X)

print(df.head(20))
```
